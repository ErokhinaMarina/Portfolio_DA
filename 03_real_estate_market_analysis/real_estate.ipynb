# Исследование объявлений о продаже квартир

## Описание проекта

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

**Цель исследования:**

Используя данные сервиса Яндекс.Недвижимость необходимо выявить основные параметры и факторы, влияющие на рыночную цену квартиры.

**Ход исследования:**
* обзор данных;
* предобработка данных;
* исследовательский анализ данных.

**Описание данных**

* airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
* balcony — число балконов
* ceiling_height — высота потолков (м)
* cityCenters_nearest — расстояние до центра города (м)
* days_exposition — сколько дней было размещено объявление (от публикации до снятия)
* first_day_exposition — дата публикации
* floor — этаж
* floors_total — всего этажей в доме
* is_apartment — апартаменты (булев тип)
* kitchen_area — площадь кухни в квадратных метрах (м²)
* last_price — цена на момент снятия с публикации
* living_area — жилая площадь в квадратных метрах (м²)
* locality_name — название населённого пункта
* open_plan — свободная планировка (булев тип)
* parks_around3000 — число парков в радиусе 3 км
* parks_nearest — расстояние до ближайшего парка (м)
* ponds_around3000 — число водоёмов в радиусе 3 км
* ponds_nearest — расстояние до ближайшего водоёма (м)
* rooms — число комнат
* studio — квартира-студия (булев тип)
* total_area — общая площадь квартиры в квадратных метрах (м²)
* total_images — число фотографий квартиры в объявлении

## Предобработка данных

###  Поиск и обработка пропусков

Найдем и изучим пропущенные значения в столбцах. Выведем количество пропущенных значений для каждого столбца. Для этого используем комбинацию двух методов.

```python
data.isnull().sum() 
```

Также, помимо абсолютных значений мы можем посмотреть долю пропусков, и на основании этого принять решение о удалении, замене либо игнорировании пропусков.

```python
# Посмотрим насколько заполнены столбцы
pd.DataFrame(round(data.isna().mean()*100,1)).style.background_gradient('coolwarm')
```

Возможные причины возникновения пропусков описала по каждому столбцу ниже:
* **ceiling_height** - значения в столбце могли быть пропущены из-за незнания данной информации. В данном столбце встречаются значения, сильно отличающиеся от правды(высота потолков больше 10 метров). Скорее всего при вводе информации допущена ошибка. Пропуски в столбце будем заменять медианным значением. 
* **floors_total** - значения могут быть пропущены из-за того, что в доме всего один этаж, либо эту информацию просто пропустили. Так как таких пропусков немного, то можно будет удалить эти строки. 
* **living_area** - значения в столбце могли быть пропущены из-за незнания данной информации. Данные пропуски оставляем без изменений.
* **is_apartment** - значения в столбце могли быть пропущены из-за того, что люди не указывали эту информацию, так как по умолчанию считали, что квартира не является аппартаментами, либо не знали эту информацию. Можем заменить пропуски на False.
* **kitchen_area** -значения в столбце могли быть пропущены из-за незнания данной информации. Данные пропуски оставляем без изменений.
* **balcony** - отсутствие данных по количеству балконов скорее всего говорит о том, что балконов в квартире нет.

#### Тип этажа квартиры

Добавим в таблицу новый столбец categorize_floor, в котором укажем тип этажа квартиры (значения — «‎первый», «последний», «другой»).

```python
#Используем функцию с именем categorize_floor() и метод apply()
def categorize_floor(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    else:
        return 'другой'

data['categorize_floor'] = data.apply(categorize_floor,axis = 1)
display(data.head(20))
```

```python
# Построим график для столбца days_exposition с границами от 0 до 230
data.plot(y = 'days_exposition', range = (0,230), bins = 100, grid = True, kind = 'hist', figsize = (7,5));
```

На графике есть ярко выраженные пики, находящиеся в районе 50 дней. 


Возвращаясь к описательной статистике столбца days_exposition определяем что среднее значение 180 дней, медиана 95 дней. Среднее значение почти в два раза больше медианы. Стандартное отклонение больше среднего значения. Такие данные могут говорить о том, что в столбце days_exposition много выбросов. Подтвердим предположения диаграммой размаха. 

```python
# Построим диаграмму размаха 
data.boxplot(column='days_exposition', figsize=(8, 6));
```

Все, что находится выше верхнего уса - выбросы. Посчитаем их по формуле Q3 + 1.5 * IQR, где Q3 - третий квартиль (отсечка трёх четвертей: 75% элементов меньше и 25% элементов больше него) и IQR - «межквартильный размах».
Получаем: 231 + 1.5 * 186 = 510.

```python
# Избавимся от выбросов
data = data.query('days_exposition < 510')
```
