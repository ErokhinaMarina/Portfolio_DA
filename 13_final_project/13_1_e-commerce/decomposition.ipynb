# Декомпозиция

## Постановка целей и задач

**Цель**: 

Провести полноценный анализ товарного ассортимента интернет-магазина товаров для дома "Пока все еще тут" и сегментировать профили покупателей для предоставления индивидуальных предложений, с целью выработки рекомендации для заказчика по взаимодействию с клиентами для улучшения показателей магазина. 

**Задачи:**

1. Провести исследовательский анализ данных 

2. Проанализировать ассортимент, представленный в интернет-магазине

3. Разделить покупателей на сегменты

4. Сформулировать и проверить статистические гипотезы

5. Изучить сегменты и выработать рекомендации: 

* клиентам какого сегмента, когда и какие категории товара рассылать в рекламных предложениях,

* как и для какого сегмента можно попытаться увеличить количество продаваемого товара, повысить возвращаемость покупателей.

## Изучение и предобработка данных

1. Загрузка данных из файла в датафрейм и изучение общей информации

2. Проверка данных на наличие дубликатов и пропусков 

3. Проверка типов данных в каждом столбце

4. Проверка данных на выбросы по количеству заказов, стоимости и среднему чеку и их устранение  

5. Добавление в таблицу новых столбцов со следующими параметрами: стоимость товара в строке, месяц заказа.

## Исследовательский анализ данных

1. Проведение общего анализа показателей магазина:

* Посмотреть период, за который имеются данные,

* Узнать количество уникальных покупателей,

* Узнать количество заказов,

* Узнать средний чек,

* Узнать среднее количество заказов на одного покупателя,

* Узнать среднее количество единиц товара при одной покупке,

* Посмотреть самые популярные товары,

* Произвести визуализацию данных.


2. Категоризация товаров:

* Разбить товары по категориям,

* Посчитать количество товаров, средний чек каждой категории,

* Провести распределение категорий товаров по количеству покупателей,

* Посмотреть распределение продаж разных категорий товаров по месяцам.


3. Сегментация покупателей с использованием RFM-анализа.

## Формулировка и проверка статистических гипотез

1. Различие в среднем чеке между сегментами покупателей

2. Различие по частоте заказов между сегментами

##  Изучение сегментов покупателей

1. Проведение анализа показателей сегментов покупателей:

* Посчитать количество товаров в заказе, количество заказов на сегмент, средний чек,

* Посмотреть распределение продаж разных сегментов покупателей по месяцам,

* Посмотреть самые популярные и непопулярные товары в каждом сегменте

* Произвести визуализацию данных.

## Выводы и рекомендации

1. Форулировка выводов

2. Выработка рекомендаций по работе с сегментами на основании полученных результатов

3. Подготовка презентации
