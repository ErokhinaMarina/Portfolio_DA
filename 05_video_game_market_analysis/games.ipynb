# Анализ рынка продаж компьютерных игр

## Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Определить закономерности успешности игры.

### Поиск и обработка пропусков

#### Поиск пропусков

Найдем и изучим пропущенные значения в столбцах.

Выведем на экран количество пропущенных значений для каждого столбца. Для этого используем комбинацию двух методов.

```python
data.isnull().sum() 
```

Отобразим на графике процент пропусков и оценим масштаб проблемы.

```python
# пропущенные значения

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 16)
            .set_title('Процент пропусков в столбцах' + "\n",fontsize = 22, color = 'SteelBlue')
            
        );    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
```

```python
pass_value_barh(data)
```

Исходя из графика, видим, что на столбцы critic_score, user_score, rating приходится самый большой процент пропусков. 

Возможные причины возникновения пропусков и решение по их обработке описала по каждому столбцу ниже:

* **name** - нужно посмотреть эти пропуски, но так как их всего два, то лучше их удалить.
* **year_of_release** - возможно данный столбец был не обязателен для заполнения, поэтому в столбце присутствуют пропуски. Удалить эти пропуски было бы не хорошо. Посчитать среднее или медиану тут неуместно, но оставлять их так нельзя. Возможно надо заменить все пропуски на ноль.
* **genre** - такая же ситуация, как и в столбце name.
* **critic_score** - оценку могли не поставить из-за того, что игра малоизвестная. Пропуски в этом столбце мы не можем удалить, так как они занимают большой процент от всех данных, но и оставить без изменений не можем. Посчитать среднее и медиану нельзя, так как это может повлиять на данные в целом, мы же ведь не знаем какая оценка могла быть у этих игр. Можно было бы заменить пропуски на нули, но это было бы не правильно, так как нулевая оценка встречается в столбце. Можно заменить пропуски на отрицательные значения. 
* **user_score** - такая же ситуация, как и в столбце critic_score.
* **rating** - скорее всего рейтинг появился не сразу, поэтому в более ранних играх он отсутствует. Заменим пропуски на indefined.

#### Обработка значений в столбце user_score

Выведем на экран все уникальные значения столбца user_score при помощи unique().

```python
display(data['user_score'].unique())
```

В столбце есть значения nan и tbd. tbd расшифровывается как To Be Determined, и означает, что данных для оценки очень мало.

Преобазуем значения в столбце user_score в float.

Если нужно специально присвоить значение NaN элементу структуры данных, для этого используется np.NaN (или np.nan) из библиотеки numpy.

```python
data['user_score'] = data['user_score'].replace('tbd', np.NaN)
data['user_score'] = data['user_score'].astype(float)
```

Следующим шагом заменим пропуски в столбце  user_score -1 при помощи fillna().

```python
data['user_score'] = data['user_score'].fillna(-1)
```

#### Столбец year_of_release

Заменим пропуски в столбце year_of_release нулями при помощи fillna().

```python
data['year_of_release'] = data['year_of_release'].fillna(0)
```

```python
data = data.dropna(subset=['name'])
```

### Проверка данных на наличие дубликатов

Сгруппируем данные. Так как value_counts() возвращает Series, а не Dataframe, а результаты нам нужно получить в последнем виде, можно использовать функцию .to_frame(). Далее переиндексируем строки с помощью reset_index(), а затем вызовем sort_values(), как обычный фрейм данных.

При помощи .nlargest() вернем первые n строк, упорядоченных по столбцам.

Аргумент ignore_index=True указывает pandas игнорировать исходные номера индексов в каждом DataFrame и создавать новый индекс, который начинается с 0 для нового DataFrame.

Для вывода подзаголовков воспользуемся ax[i].set_title(), добавив к названию каждого графика свой регион. 

labels: параметр представляет собой последовательность строк, обеспечивающих метки для каждого сегмента.

autopct:  параметр представляет собой строку или функцию, используемую для маркировки клиньев их числовым значением.


```python
top_5_platforms = actual_period.groupby('platform')[['na_sales', 'eu_sales', 'jp_sales']].sum()
columns = top_5_platforms.columns
region = ['Северной Америке', 'Европе', 'Японии']

fig , ax = plt.subplots(nrows = 1, ncols = 3, figsize=(18,5))
fig.suptitle('Топ-5 самых популярных платформ', fontsize=22)
for i in [0, 1, 2]:
    groupby_platforms = actual_period\
        .groupby('platform')[columns[i]].sum().to_frame().reset_index()\
        .sort_values(by = columns[i], ascending=False)
    platform_5 = groupby_platforms.nlargest(5, columns[i])
    other_value = (groupby_platforms[columns[i]].sum()) - (platform_5[columns[i]].sum())
    new_row = {'platform':'другие', columns[i]:other_value}
    platform_5 = platform_5.append(new_row, ignore_index = True)
    ax[i].set_title('Популярный платформы в ' + region[i])
    ax[i].pie(platform_5[columns[i]], labels=platform_5['platform'], autopct='%1.1f%%')
plt.show()



```

**Выводы:** На графиках видно, что PS4 и XOne входят в тройку самых популярных платформа в Северной Америке и Европе.
В Японии наиболее популярна платформа 3DS.

## Проверка гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Альтернативная гипотеза : Средние пользовательские рейтинги платформ Xbox One и PC отличаются

```python
XOne_hypothesis = actual_period\
    [(actual_period['platform']=='XOne') & (actual_period['user_score'] > 0)]['user_score'].dropna()
PC_hypothesis = actual_period\
    [(actual_period['platform']=='PC') & (actual_period['user_score'] > 0)]['user_score'].dropna()
print('Средняя оценка для XOne:', XOne_hypothesis.mean())
print('Средняя оценка для PC:', PC_hypothesis.mean())
```

Так как нам необходимо сравнивать средние двух генеральных совокупностей между собой, используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind().
Использовать данный метод мы можем, так как:

* Генеральные совокупности не зависят друг от друга

* Выборочные средние нормально распределены

```python
alpha = .05

results = st.ttest_ind( XOne_hypothesis, PC_hypothesis, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Выводы:** Значение p-value: 0.14759594013430463. Так как значение p-value больше значения alpha, то, следует, что отвергнуть нулевую гипотезу мы не можем.


### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports одинаковые

Альтернативная гипотеза : Средние пользовательские рейтинги жанров Action и Sports отличаются

```python
action_hypothesis = actual_period\
    [(actual_period['genre'] == 'Action') &(actual_period['user_score'] > 0)]['user_score'].dropna()
sports_hypothesis = actual_period\
    [(actual_period['genre'] == 'Sports') &(actual_period['user_score'] > 0)]['user_score'].dropna()
print('Средняя оценка для action:', action_hypothesis.mean())
print('Средняя оценка для sports:', sports_hypothesis.mean())
```

Так как нам необходимо сравнивать средние двух генеральных совокупностей между собой, используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind().
Использовать данный метод мы можем, так как:

* Генеральные совокупности не зависят друг от друга

* Выборочные средние нормально распределены

```python
alpha = .05

result = st.ttest_ind(action_hypothesis, sports_hypothesis, equal_var=False)

print('p-значение:', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Выводы:** На основании того, что значение p-value: 1.4460039700704315e-20, что меньше нашего уровня статистической значимости, мы можем опровергнуть нулевую гипотезу. А это значит, что средние пользовательские рейтинги жанров Action и Sports разные.
